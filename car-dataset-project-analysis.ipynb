```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
```

```python
# Load the dataset
file_path = '/kaggle/input/all-cars-in-india-price-and-specifications/All_cars_dataset.csv'
df = pd.read_csv(file_path)

pd.DataFrame(df)
```

```python
# Display the first few rows and the summary of the dataset
df.head()
```

```python
df.describe()
```

```python
# Fill missing values in the 'ENGINE' column
df['ENGINE'] = df['ENGINE'].fillna(df['ENGINE'].mode()[0])

#  Convert all values in the 'Price' column to strings and then clean
df['Price'] = df['Price'].astype(str)  # Ensure all values are strings
df['Price'] = df['Price'].str.replace('Lakh', '').str.replace('Crore', '').str.replace(',', '')
df['Price'] = df['Price'].astype(float)  # Convert to float after cleaning

#  Clean the 'EMI' column (also converting to string first)
df['EMI'] = df['EMI'].astype(str).str.replace(',', '').astype(float)

#  Clean the 'Mileage' and 'Top_Speed' columns (convert to strings first)
df['Mileage'] = df['Mileage'].astype(str).str.extract('(\d+)').astype(float)
df['Top_Speed'] = df['Top_Speed'].astype(str).str.extract('(\d+)').astype(float)

#  Convert some columns to categories
df['TRANSMISSION'] = df['TRANSMISSION'].astype('category')
df['FUEL TYPE'] = df['FUEL TYPE'].astype('category')
df['Colors'] = df['Colors'].astype('category')

#  Remove duplicate rows
df = df.drop_duplicates()


```

```python
# Display cleaned data info
df.info()
```

# SOME OPERATIONS : PANDAS AND NUMPY
Step 1: Basic Summary of the Dataset
We’ll start by getting an overview of the dataset, such as the total number of cars, average values, and counts of unique categories.

Step 2: Descriptive Statistics
We’ll use Pandas to calculate descriptive statistics, such as mean, median, mode, and standard deviation for numeric columns.

Step 3: Grouping and Aggregation
We’ll explore how different groups (e.g., based on FUEL TYPE, TRANSMISSION) perform in terms of price and mileage.

Step 4: Correlations
We’ll calculate correlations between numeric columns to see how they are related.

Here’s the code to perform these analyses:

```python

# Step 1: Basic Summary of the Dataset

# Total number of cars
total_cars = len(df)
print(f"Total number of cars: {total_cars}")

# Average Price, Mileage, and Top Speed
avg_price = df['Price'].mean()
avg_mileage = df['Mileage'].mean()
avg_top_speed = df['Top_Speed'].mean()

print(f"Average Price: {avg_price} Lakhs")
print(f"Average Mileage: {avg_mileage} kmpl")
print(f"Average Top Speed: {avg_top_speed} km/h")

# Count of different Transmission types and Fuel types
transmission_counts = df['TRANSMISSION'].value_counts().to_frame()
fuel_type_counts = df['FUEL TYPE'].value_counts().to_frame()

print("\nNumber of cars by Transmission Type:\n", transmission_counts)
print("\nNumber of cars by Fuel Type:\n", fuel_type_counts)


```

```python
# Step 2: Descriptive Statistics

# Basic statistics for numeric columns
basic_stats = df[['Price', 'Mileage', 'Top_Speed']].describe()
#print(basic_stats)
pd.DataFrame(basic_stats)
```

```python
# Step 3: Grouping and Aggregation

# Average Price for each Fuel Type
avg_price_by_fuel = df.groupby('FUEL TYPE', observed=True)['Price'].mean().reset_index()
#print("\nAverage Price by Fuel Type:\n", avg_price_by_fuel)
pd.DataFrame(avg_price_by_fuel)
```

```python
# Average Mileage for each Transmission Type
avg_mileage_by_transmission = df.groupby('TRANSMISSION', observed=True)['Mileage'].mean().reset_index()
#print("\nAverage Mileage by Transmission Type:\n", avg_mileage_by_transmission)
pd.DataFrame(avg_mileage_by_transmission)
```

```python

# Step 4: Correlations

# Correlation between Price, Mileage, and Top Speed
correlations = df[['Price', 'Mileage', 'Top_Speed']].corr().reset_index()
#print("\nCorrelation between Price, Mileage, and Top Speed:\n", correlations)
pd.DataFrame(correlations)
```

```python
#Task 1: Find the Most and Least Expensive Cars
# Most expensive car
most_expensive_car = df.loc[df['Price'].idxmax(), ['Name', 'Price']]

# Least expensive car
least_expensive_car = df.loc[df['Price'].idxmin(), ['Name', 'Price']]

#print("Most Expensive Car:", most_expensive_car)
pd.DataFrame(most_expensive_car)
```

```python
#print("Least Expensive Car:", least_expensive_car)
pd.DataFrame(least_expensive_car)
```

```python
# Task 2: Analyze the Relationship Between Engine Size and Top Speed

# Average top speed by engine size
avg_top_speed_by_engine = df.groupby('Engine_Size')['Top_Speed'].mean().reset_index()

pd.DataFrame(avg_top_speed_by_engine)
```

```python
# Average mileage by fuel type

avg_mileage_by_fuel = df.groupby('FUEL TYPE', observed=True)['Mileage'].mean().reset_index()


pd.DataFrame(avg_mileage_by_fuel)
```

```python
# Summary of average price, mileage, and top speed by fuel type and transmission
summary_table = df.groupby(['FUEL TYPE', 'TRANSMISSION'], observed=True).agg({
    'Price': 'mean',
    'Mileage': 'mean',
    'Top_Speed': 'mean'
}).reset_index()

pd.DataFrame(summary_table)
```

# 1. Company Comparison
* Compare average car prices, mileage, and top speed by company.
* Identify the companies with the most and least expensive cars.
* Analyze the distribution of different car types (e.g., sedan, SUV) by company if the data allows

```python
# Extract company names from car names
df['Company'] = df['Name'].apply(lambda x: x.split()[0])

# Group by company and calculate average price, mileage, and top speed
company_comparison = df.groupby('Company').agg({
    'Price': 'mean',
    'Mileage': 'mean',
    'Top_Speed': 'mean'
}).reset_index()

# Display the top companies by average price
company_comparison = company_comparison.sort_values(by='Price', ascending=False)
company_comparison
```

```python
import matplotlib.pyplot as plt

# 1. Company Comparison - Average Price
plt.figure(figsize=(12, 6))
plt.bar(company_comparison['Company'], company_comparison['Price'], color='blue')
plt.title('Average Price by Company')
plt.xlabel('Company')
plt.ylabel('Average Price (in Lakhs)')
plt.xticks(rotation=90)
plt.show()
```

# Step-by-Step Analysis
* Step 1: Calculate Performance Index
* We'll create a Performance Index by combining Top_Speed and Engine_Size. A higher index indicates better performance.
* 
* Step 2: Calculate Price-to-Performance Ratio
* We'll then calculate the Price-to-Performance Ratio for each brand. A lower ratio indicates better value for money in terms of performance.
* 
* Step 3: Identify Top Brands
* We'll identify the top 5 brands that offer the best value based on the Price-to-Performance Ratio.


```python
1 # Normalize Top_Speed and Engine_Size to create a balanced Performance Index
df['Performance_Index'] = (df['Top_Speed'] / df['Top_Speed'].max()) + (df['Engine_Size'] / df['Engine_Size'].max())
df
```

```python
2# Group by brand to calculate the average Price-to-Performance Ratio
brand_performance = df.groupby('Brand').agg({
    'Price': 'mean',
    'Performance_Index': 'mean'
}).reset_index()

# Calculate the Price-to-Performance Ratio
brand_performance['Price_to_Performance'] = brand_performance['Price'] / brand_performance['Performance_Index']

# Sort brands by the Price-to-Performance Ratio
best_value_brands = brand_performance.sort_values(by='Price_to_Performance')

pd.DataFrame(best_value_brands )
```

```python
# 3. Visualize the Top Brands



import matplotlib.pyplot as plt

# Visualize the top 5 brands by Price-to-Performance Ratio
plt.figure(figsize=(12, 6))
plt.barh(best_value_brands['Brand'].head(5), best_value_brands['Price_to_Performance'].head(5), color='teal')
plt.title('Top 5 Brands by Price-to-Performance Ratio')
plt.xlabel('Price-to-Performance Ratio')
plt.ylabel('Brand')
plt.gca().invert_yaxis()  # Invert y-axis to have the best value at the top
plt.show()

```
